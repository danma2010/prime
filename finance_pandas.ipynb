```python
import pandas_datareader.data as web
import datetime
```

```python
import matplotlib.pyplot as plt
```

```python
start = datetime.datetime(2020,1,1)
end = datetime.datetime(2021,6,1)
print(start)

google = web.DataReader("GOOGL",'yahoo', start, end)
```

```python
google.head
```

```python

google['Open'].plot(label = 'GOOGL open', figsize=(15,7))
google['Close'].plot(label = 'GOOGL close')
plt.legend()
plt.title('GOOGL')
plt.ylabel('price')
plt.show()

google['Volume'].plot(figsize=(15,4))
plt.ylabel('Volume')
plt.show()
```

```python
import numpy as np
%matplotlib inline
```

```python
tsla = web.DataReader("TSLA",'yahoo', start, end)
ford = web.DataReader("F",'yahoo', start, end)
gm = web.DataReader("GM",'yahoo', start, end)
```

```python
ford.head
```

```python
tsla.to_csv('tsla.csv')
ford.to_csv('ford.csv')
gm.to_csv('gm.csv')
```

```python
tsla['Close'].plot(label = 'TSLA open', figsize=(15,7))
ford['Close'].plot(label = 'F open', figsize=(15,7))
gm['Close'].plot(label = 'GM open', figsize=(15,7))
plt.legend()
#plt.title('GOOGL')
plt.ylabel('price')
```

```python
type(ford)
```

```python
tsla['TotalTrad'] = tsla['Close'] * tsla['Volume']/1000000
ford['TotalTrad'] = ford['Close'] * ford['Volume']/1000000
gm['TotalTrad'] = gm['Close'] * gm['Volume']/1000000
```

```python
tsla['TotalTrad'].plot(label = 'TSLA open', figsize=(15,7))
ford['TotalTrad'].plot(label = 'F open', figsize=(15,7))
gm['TotalTrad'].plot(label = 'GM open', figsize=(15,7))
plt.legend()
#plt.title('GOOGL')
plt.ylabel('TotalTrad')
plt.show()
```
